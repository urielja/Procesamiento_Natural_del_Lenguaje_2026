En este notebook utilizaremos el framework Trax y aprenderemos sobre algunos de sus bloques de construcción básicos.



### ¿Por qué Trax y no TensorFlow o PyTorch?

TensorFlow y PyTorch son frameworks extensos que pueden hacer casi cualquier cosa en deep learning. Ofrecen mucha flexibilidad, pero eso a menudo implica una sintaxis verbosa y tiempo extra para programar.

Trax es mucho más conciso. Se ejecuta sobre un backend de TensorFlow pero permite entrenar modelos con comandos de 1 línea. Trax también funciona de extremo a extremo, permitiéndote obtener datos, modelar y entrenar todo con instrucciones únicas y breves. Esto significa que puedes enfocarte en aprender, en lugar de gastar horas en las peculiaridades de la implementación de un framework grande.

### Cómo programar en Trax
Construir modelos en Trax se basa en 2 conceptos clave: **capas (layers)** y **combinadores (combinators)**.
Las capas de Trax son objetos simples que procesan datos y realizan cálculos. Pueden encadenarse para formar capas compuestas usando los combinadores de Trax, permitiéndote construir capas y modelos de cualquier complejidad.

### Trax, JAX, TensorFlow y Tensor2Tensor

Ya sabemos que Trax usa Tensorflow como backend, pero también usa la librería JAX para acelerar el cálculo. Podemos ver JAX como una versión mejorada y optimizada de numpy.

Como resultado de esto, donde antes encontramos el tipo `numpy.ndarray` ahora encontraremos el tipo `jax.interpreters.xla.DeviceArray`.

Tensor2Tensor es otro nombre que quizás hayas escuchado. Comenzó como una solución de extremo a extremo muy parecida a como está diseñado Trax, pero creció de forma difícil de manejar y complicada. Así que puedes ver a Trax como la nueva versión mejorada que opera de forma mucho más rápida y simple.

### Capa Concatenate


````python
# Crear una capa concatenate de trax
concat = tl.Concatenate()
print("-- Propiedades --")
print("name :", concat.name)
print("expected inputs :", concat.n_in)
print("promised outputs :", concat.n_out, "\n")

# Entradas
x1 = np.array([-10, -20, -30])
x2 = x1 / -10
print("-- Entradas --")
print("x1 :", x1)
print("x2 :", x2, "\n")
````

## Las capas son configurables
Se puede cambiar la configuración predeterminada de las capas. Por ejemplo, se puede cambiar las entradas esperadas de una capa concatenate de 2 a 3 usando el parámetro opcional `n_items`.

````python
# Configurar una capa concatenate
concat_3 = tl.Concatenate(n_items=3)  # configurar las entradas esperadas de la capa
print("-- Propiedades --")
print("name :", concat_3.name)
print("expected inputs :", concat_3.n_in)
print("promised outputs :", concat_3.n_out, "\n")
````

````python
# Entradas
x1 = np.array([-10, -20, -30])
x2 = x1 / -10
x3 = x2 * 0.99
print("-- Entradas --")
print("x1 :", x1)
print("x2 :", x2)
print("x3 :", x3, "\n")
````

````python
# Salidas
y = concat_3([x1, x2, x3])
print("-- Salidas --")
print("y :", y)
````

## Las capas pueden tener pesos
Algunos tipos de capas incluyen pesos y sesgos mutables que se usan en el cálculo y el entrenamiento. Las capas de este tipo requieren inicialización antes de usarse.

Por ejemplo, la capa `LayerNorm` calcula datos normalizados, que también se escalan mediante pesos y sesgos. Durante la inicialización pasas la forma (shape) y el tipo de dato de las entradas, para que la capa pueda inicializar arreglos compatibles de pesos y sesgos.

````python
# help(tl.LayerNorm)
# help(shapes.signature)
````

````python
# Inicialización de la capa
norm = tl.LayerNorm()
# Primero debemos saber cómo se verán los datos de entrada
x = np.array([0, 1, 2, 3], dtype="float")

# Usar la firma (signature) de los datos de entrada para obtener la forma y el tipo para inicializar pesos y sesgos
norm.init(shapes.signature(x)) # Necesitamos convertir el tipo de dato de entrada de la tupla habitual a ShapeDtype de trax
````

````python
print("Normal shape:",x.shape, "Data Type:",type(x.shape))
print("Shapes Trax:",shapes.signature(x),"Data Type:",type(shapes.signature(x)))
````

````python
# Inspeccionar propiedades
print("-- Propiedades --")
print("name :", norm.name)
print("expected inputs :", norm.n_in)
print("promised outputs :", norm.n_out)
# Pesos y sesgos
print("weights :", norm.weights[0])
print("biases :", norm.weights[1], "\n")

# Entradas
print("-- Entradas --")
print("x :", x)

# Salidas
y = norm(x)
print("-- Salidas --")
print("y :", y)
````

## Capas personalizadas

También podemos crear nuestras propias capas personalizadas y definir funciones personalizadas para los cálculos usando `tl.Fn`. 

````python
#help(tl.Fn)
````

````python
# Definir una capa personalizada
# Este ejemplo creaa una capa para calcular la entrada multiplicada por 2
def TimesTwo():
    layer_name = "TimesTwo" #no olvides darle un nombre a tu capa personalizada para identificarla

    # Función personalizada para la capa personalizada
    def func(x):
        return x * 2

    return tl.Fn(layer_name, func)

````

````python
# Probarla
times_two = TimesTwo()
````

````python
# Inspeccionar propiedades
print("-- Propiedades --")
print("name :", times_two.name)
print("expected inputs :", times_two.n_in)
print("promised outputs :", times_two.n_out, "\n")

# Entradas
x = np.array([1, 2, 3])
print("-- Entradas --")
print("x :", x, "\n")

# Salidas
y = times_two(x)
print("-- Salidas --")
print("y :", y)
````

## Combinadores (Combinators)
Podemos combinar capas para construir capas más complejas. Trax proporciona un conjunto de objetos llamados capas combinadoras para lograr esto. Los combinadores son ellos mismos capas, por lo que el comportamiento es conmutable.



### Combinador Serial
Este es el más común y fácil de usar. Por ejemplo, podríamos construir una red neuronal simple combinando capas en una sola capa usando el combinador `Serial`. Esta nueva capa luego actúa como una sola capa, así que podemos inspeccionar entradas, salidas y pesos. O incluso combinarla en otra capa

Los combinadores pueden entonces usarse como modelos entrenables.

**Nota: si existe un combinador serial, también debe existir un combinador paralelo.**


````python
# help(tl.Serial)
# help(tl.Parallel)
````

````python
# Combinador serial
serial = tl.Serial(
    tl.LayerNorm(),         # normalizar la entrada
    tl.Relu(),              # convertir los valores negativos a cero
    times_two,              # la capa personalizada que creamos arriba, multiplica por 2 la entrada recibida de arriba    
)
````

````python
# Inicialización
x = np.array([-2, -1, 0, 1, 2]) #entrada
serial.init(shapes.signature(x)) #inicializando la instancia serial
````

````python
print("-- Modelo Serial --")
print(serial,"\n")
print("-- Propiedades --")
print("name :", serial.name)
print("sublayers :", serial.sublayers)
print("expected inputs :", serial.n_in)
print("promised outputs :", serial.n_out)
print("weights & biases:", serial.weights, "\n")
````

````python
# Entradas
print("-- Entradas --")
print("x :", x, "\n")
````

````python
# Salidas
y = serial(x)
print("-- Salidas --")
print("y :", y)
````

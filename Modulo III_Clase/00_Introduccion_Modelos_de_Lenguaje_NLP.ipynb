## ¿Qué es un Modelo de Lenguaje?

Un **Modelo de Lenguaje** (en inglés, *Language Model* o **LM**) es un modelo de probabilidad sobre secuencias de palabras. Dicho de forma sencilla:

> Un modelo de lenguaje **asigna una probabilidad** a una secuencia de palabras y, gracias a ello, puede **predecir cuál es la palabra más probable que viene a continuación**.

Por ejemplo, ante la frase incompleta:

> *"El cielo está..."*

un buen modelo de lenguaje asignará una probabilidad alta a palabras como **"azul"**, **"nublado"** o **"despejado"**, y una probabilidad muy baja a palabras sin sentido en ese contexto como **"zapato"** o **"comiendo"**.


## Intuición: predecir la siguiente palabra

El **autocompletado** del teclado de tu celular o de la barra de búsqueda de Google es el ejemplo más cotidiano de un modelo de lenguaje en acción.

Cuando escribes:

> *"¿Cómo estás..."*

el teclado te sugiere **"hoy"**, **"tú"** o **"?"** porque ha aprendido, a partir de millones de frases, qué palabras suelen seguir.

La idea clave es que el lenguaje tiene **estructura y regularidades**. Algunas secuencias de palabras son mucho más frecuentes y "naturales" que otras, y el modelo aprende esas regularidades a partir de grandes cantidades de texto.

## Formalización matemática

Dado una secuencia de palabras $w_1, w_2, \dots, w_n$, un modelo de lenguaje calcula la probabilidad conjunta de esa secuencia:

$$ P(w_1, w_2, \dots, w_n) $$

Usando la **regla de la cadena** de la probabilidad, esta se descompone en un producto de probabilidades condicionales:

$$ P(w_1, w_2, \dots, w_n) = \prod_{t=1}^{n} P(w_t \mid w_1, w_2, \dots, w_{t-1}) $$

Es decir, la probabilidad de cada palabra depende de **todas las palabras anteriores**.

El problema es que considerar *todo* el historial es computacionalmente inviable. Por eso se hace una **suposición de Markov**: la probabilidad de la siguiente palabra depende solo de las últimas $n-1$ palabras (esto da lugar a los **n-gramas**):

$$ P(w_t \mid w_1, \dots, w_{t-1}) \approx P(w_t \mid w_{t-n+1}, \dots, w_{t-1}) $$

Por ejemplo, un modelo de **bigramas** (n=2) solo mira la palabra inmediatamente anterior:

$$ P(\text{"azul"} \mid \text{"está"}) $$

````python
# Ejemplo conceptual: estimar P(palabra | palabra_anterior) contando frecuencias
from collections import Counter

corpus = "el cielo esta azul el cielo esta despejado el mar esta azul".split()

# Cuenta los bigramas (pares de palabras consecutivas)
bigramas = Counter(zip(corpus[:-1], corpus[1:]))
# Cuenta las palabras individuales (unigramas)
unigramas = Counter(corpus)

# Estima P('azul' | 'esta') = conteo('esta azul') / conteo('esta')
p = bigramas[('esta', 'azul')] / unigramas['esta']
print(f"P('azul' | 'esta') = {p:.3f}")
````

## Tipos de modelos de lenguaje

Los modelos de lenguaje han evolucionado enormemente. Podemos agruparlos en tres grandes familias:

<a name='4.1'></a>
### Modelos estadísticos (N-gramas)

Son los modelos **clásicos**. Estiman las probabilidades simplemente **contando frecuencias** de secuencias de palabras en un corpus de entrenamiento.

- ✅ Sencillos, rápidos e interpretables.
- ❌ No capturan relaciones de largo alcance ni significado semántico.
- ❌ Sufren del problema de **dispersión** (muchas combinaciones nunca aparecen → probabilidad cero). Se resuelve parcialmente con técnicas de **suavizado** (Laplace, Kneser-Ney).

[Speech and Language Processing — Jurafsky & Martin, Capítulo de N-gramas](https://web.stanford.edu/~jurafsky/slp3/3.pdf).*

<a name='4.2'></a>
### Modelos neuronales

En lugar de contar, estos modelos **aprenden representaciones vectoriales** de las palabras (los famosos *word embeddings* como **Word2Vec** o **GloVe**) y usan redes neuronales (RNN, LSTM, GRU) para predecir la siguiente palabra.

- ✅ Capturan **significado semántico**: palabras parecidas tienen vectores parecidos.
- ✅ Manejan mejor secuencias largas que los n-gramas.
- ❌ Más costosos de entrenar.

<img src="https://jalammar.github.io/images/word2vec/word2vec.png" width="550"/>

*Imagen: los embeddings representan palabras como vectores en un espacio continuo. Fuente: [Jay Alammar](https://jalammar.github.io/illustrated-word2vec/).*

<a name='4.3'></a>
### Transformers y Grandes Modelos de Lenguaje (LLMs)

La arquitectura **Transformer** (2017, paper *[Attention is All You Need](https://arxiv.org/abs/1706.03762)*) revolucionó el NLP gracias al mecanismo de **atención** (*attention*), que permite al modelo "prestar atención" a las palabras más relevantes del contexto, sin importar qué tan lejos estén.

De aquí nacen los **Grandes Modelos de Lenguaje (LLMs)** como **GPT**, **BERT**, **LLaMA** o **Gemini**, entrenados con cantidades masivas de texto.

- ✅ Capturan contexto de **muy largo alcance**.
- ✅ Capaces de traducir, resumir, responder preguntas, generar código, etc.
- ❌ Requieren enormes recursos de cómputo y datos.

## ¿Cómo se evalúan? La perplejidad

La métrica más común para evaluar un modelo de lenguaje es la **perplejidad** (*perplexity*, PP). Mide qué tan "sorprendido" queda el modelo al ver un texto nuevo.

$$ PP(W) = \sqrt[N]{ \prod_{t=1}^{N} \frac{1}{P(w_t \mid w_{1} \cdots w_{t-1})} } $$

- Una perplejidad **baja** significa que el modelo predice bien el texto (está poco "perplejo").
- Una perplejidad **alta** significa que el modelo se sorprende mucho (predice mal).

Intuitivamente, la perplejidad es el número promedio de palabras entre las que el modelo "duda" en cada paso. Un modelo con perplejidad 10 es como si en cada palabra eligiera al azar entre 10 opciones igualmente probables.

## Línea del tiempo de la evolución

```
1948  ──  Claude Shannon introduce los n-gramas (Teoría de la Información)
  │
1990s ──  Modelos estadísticos de n-gramas dominan el NLP
  │
2003  ──  Bengio propone el primer modelo de lenguaje neuronal
  │
2013  ──  Word2Vec (Mikolov) populariza los word embeddings
  │
2017  ──  Transformer: "Attention is All You Need"
  │
2018  ──  BERT y GPT inician la era de los LLMs
  │
2020+ ──  GPT-3, GPT-4, LLaMA, Gemini... modelos masivos
```

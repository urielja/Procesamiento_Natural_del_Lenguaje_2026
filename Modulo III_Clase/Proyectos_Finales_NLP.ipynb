## Cómo elegir tu proyecto

Considera estos criterios al seleccionar:

1. **Interés personal:** elige un tema que te motive.
2. **Disponibilidad de datos:** verifica que el dataset sea accesible y de tamaño manejable.
3. **Cobertura del curso:** el proyecto ideal combina técnicas de **al menos dos módulos** (por ejemplo, preprocesamiento clásico + un modelo Transformer).
5. **Despliegue:** se valora una demo funcional (Gradio o FastAPI)

## Proyecto 1: Detector de Noticias Falsas (Fake News) 

**Descripción:** clasifica titulares o artículos como **reales** o **falsos**. Proyecto muy atractivo por su relevancia social.

**Técnicas del curso:**
- Preprocesamiento + regex para limpiar HTML/URLs.
- TF-IDF + embeddings.
- Clasificación binaria con ML clásico y/o Transformers.
- Análisis de palabras más discriminativas.

**Datasets:**
- 📰 [Fake and Real News (Kaggle)](https://www.kaggle.com/datasets/clmentbisaillon/fake-and-real-news-dataset).
- 📰 [LIAR Dataset](https://www.cs.ucsb.edu/~william/data/liar_dataset.zip) — declaraciones políticas etiquetadas.
- 🇪🇸 [Spanish Fake News Corpus](https://github.com/jpposadas/FakeNewsCorpusSpanish) — en español.

<a name='p4'></a>
## Proyecto 2: Sistema de Autocompletado / Generación de Texto 

**Descripción:** construye un sistema que **predice la siguiente palabra** o genera texto al estilo de un autor/corpus. Conecta directamente con los notebooks de modelos de lenguaje del Módulo II.

**Técnicas del curso:**
- Modelos de **n-gramas** con suavizado y **perplejidad** (Módulo II).
- Implementación con **NLTK** (`11_Modelos_de_Lenguaje_NLTK_Autocompletado`).
- *Avanzado:* generación con **GPT-2 / Transformers** (Módulo III).
- Despliegue como teclado predictivo en Gradio.

**Datasets / Corpus:**
- 📜 [Project Gutenberg](https://www.gutenberg.org/) — libros de dominio público (¡genera texto al estilo de Cervantes o Shakespeare!).
- 📰 [WikiText-2 / WikiText-103](https://huggingface.co/datasets/wikitext).


## Proyecto 3: Traductor Automático (Español <--> Inglés)

**Descripción:** implementa un traductor, desde el enfoque **ingenuo con embeddings bilingües** hasta un modelo **seq2seq con atención**.

**Técnicas del curso:**
- **Traducción automática ingenua** con matrices de transformación de embeddings (Módulo II, `01_Naive_Machine_Translation`).
- Embeddings Word2Vec/GloVe en dos idiomas (Módulo I).
- **Mecanismo de atención** y Transformers (Módulo III).

**Datasets:**
- 🌐 [Tatoeba — pares de oraciones](https://tatoeba.org/en/downloads).
- 🌐 [ManyThings — Spa-Eng](http://www.manythings.org/anki/) — `spa-eng.zip`, ideal para empezar.
- 🌐 [OPUS — corpus paralelos](https://opus.nlpl.eu/).

## Proyecto 4: Resumen Automático de Textos (Summarization)

**Descripción:** genera resúmenes de artículos o noticias. Puedes hacer resumen **extractivo** (selección de frases con TF-IDF) o **abstractivo** (con Transformers).

**Técnicas del curso:**
- TF-IDF y ranking de oraciones para resumen extractivo.
- Embeddings para similitud entre frases.
- Modelos preentrenados **BART / T5 / mT5** para resumen abstractivo.

**Datasets:**
- 📰 [CNN/DailyMail](https://huggingface.co/datasets/cnn_dailymail) — noticias + resúmenes.
- 📰 [XSum](https://huggingface.co/datasets/EdinburghNLP/xsum).
- 🇪🇸 [MLSUM (incluye español)](https://huggingface.co/datasets/mlsum).

## Proyecto 5: Buscador Semántico de Documentos  
**Descripción:** sistema que, dada una consulta en lenguaje natural, devuelve los documentos **más relevantes por significado** (no solo por palabras exactas). Base de los sistemas RAG modernos.

**Técnicas del curso:**
- Similitud coseno entre embeddings.
- TF-IDF como baseline.
- **Sentence-Transformers** para embeddings contextuales.
- Búsqueda vectorial (FAISS) y demo en Gradio.

**Datasets:**
- 📚 [Wikipedia dumps](https://huggingface.co/datasets/wikipedia).
- ❓ [MS MARCO](https://microsoft.github.io/msmarco/) — consultas y pasajes.
- 📄 [arXiv abstracts (Kaggle)](https://www.kaggle.com/datasets/Cornell-University/arxiv).

## Proyecto 6: Detección de Spam / Toxicidad

**Descripción:** clasifica mensajes como **spam/no spam** o detecta **comentarios tóxicos**. Excelente proyecto introductorio con alto impacto práctico.

**Técnicas del curso:**
- Limpieza y normalización de texto (Módulo I).
- BoW / TF-IDF + Naive Bayes (Módulos I y II).
- Manejo de clases desbalanceadas y métricas (precisión, recall, F1).

**Datasets:**
- ✉️ [SMS Spam Collection (UCI)](https://archive.ics.uci.edu/dataset/228/sms+spam+collection).
- ☠️ [Jigsaw Toxic Comments (Kaggle)](https://www.kaggle.com/c/jigsaw-toxic-comment-classification-challenge).
- 📧 [Enron Spam](https://www2.aueb.gr/users/ion/data/enron-spam/).

## Proyecto 7: Chatbot de Preguntas y Respuestas (QA) 

**Descripción:** sistema que responde preguntas sobre un texto/documento dado (QA extractivo) o un asistente conversacional sobre un dominio específico.

**Técnicas del curso:**
- Embeddings y recuperación de contexto (Módulo I).
- **BERT/BETO para QA** (Módulo III, `02_bert_roberta`, `03_fine_tuning`).
- Despliegue completo con **FastAPI + Gradio** (Módulo III, `04_proyecto_despliegue`).

**Datasets:**
- ❓ [SQuAD 2.0](https://rajpurkar.github.io/SQuAD-explorer/).
- 🇪🇸 [SQAC — Spanish Question Answering Corpus](https://huggingface.co/datasets/PlanTL-GOB-ES/SQAC).
- 🇪🇸 [MLQA (multilingüe, incluye español)](https://huggingface.co/datasets/mlqa).

## Proyecto 8: Clasificación Multimodal (Texto + Imágenes)

**Descripción:** proyecto avanzado que combina texto e imágenes, por ejemplo: **descripción automática de imágenes** (*image captioning*), clasificación de memes, o búsqueda de imágenes por texto con CLIP.

**Técnicas del curso:**

## Recomendaciones finales

**Empieza simple:** primero un *baseline* (TF-IDF + modelo clásico) y luego mejora con Transformers. Así tendrás siempre algo que mostrar.

**Documenta tu pipeline:** datos → preprocesamiento → modelo → evaluación → despliegue.

**Mide bien:** usa métricas adecuadas (accuracy, F1, matriz de confusión, perplejidad) y conjuntos de *train/validation/test*.

**Despliega tu demo:** una interfaz en **Gradio** o una API en **FastAPI** eleva mucho la calidad del proyecto.

**Reutiliza el material del curso:** muchos notebooks (`NLP_TextCleaner.py`, pipelines de clasificación, modelos de lenguaje) son directamente reutilizables.

## Clasificación de Texto

La **clasificación de texto** es la tarea de asignar automáticamente una o más categorías predefinidas a un fragmento de texto.

#### Tipos de clasificación

| Tipo | Descripción | Ejemplo |
|---|---|---|
| **Binaria** | 2 clases | Spam / No-spam, Positivo / Negativo |
| **Multi-clase** | N clases mutuamente excluyentes | Categoría de noticia |
| **Multi-etiqueta** | Múltiples clases posibles | Tags de un artículo |

#### Aplicaciones reales

```
📧 Filtro de spam           → Spam / Ham
😊 Análisis de sentimientos  → Positivo / Negativo / Neutro
📰 Categorización de noticias → Política / Deporte / Tecnología / ...
🔍 Detección de idioma       → Español / Inglés / Francés / ...
🚨 Detección de toxicidad    → Tóxico / No tóxico
⚖️ Clasificación legal       → Tipo de contrato / cláusula
```

### Pipeline general

```
Texto crudo
    ↓
Preprocesamiento (limpieza, tokenización, stopwords)
    ↓
Representación vectorial (BoW, TF-IDF, Embeddings)
    ↓
Modelo de clasificación (Naive Bayes, SVM, BERT...)
    ↓
Predicción + Evaluación
```

## Análisis de Sentimientos con scikit-learn

### Multinomial Naive Bayes
Ahora usamos el pipeline de `scikit-learn` que combina **TF-IDF** + **Multinomial Naive Bayes** de forma eficiente.

- [Pipeline de sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html)

- [scikit-learn-pipelines guide](https://mahmoudyusof.github.io/blog/scikit-learn-pipelines)

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    textos, etiquetas, test_size=0.3, random_state=42, stratify=etiquetas
)
````

````python
# Pipeline TF-IDF + Naive Bayes 
pipeline_nb = Pipeline([('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df = 1, max_features=5000, sublinear_tf = True)),
                         ('clf', MultinomialNB(alpha=0.5))
                       ])

````

### Regresión Logística

La **Regresión Logística** aprende pesos para cada feature (palabra) usando la función sigmoide:

$$P(y=1 \mid x) = \sigma(w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$$


````python
from sklearn.linear_model import LogisticRegression
#Pipeline con LogisticRegression
pipeline_lr = Pipeline([('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df = 1, max_features=5000, sublinear_tf = True)),
                        ('clr', LogisticRegression(C = 1.0, max_iter=1000, solver = 'lbfgs'))
    
]) 
````

````python
#Entrenamoento LR
pipeline_lr.fit(X_train, y_train)
````

````python
X_test
````

````python
y_hat_test = pipeline_lr.predict(X_test) #predicción de las reseñas de testeo
````

````python
# Coeficientes: importancia de cada feature
pipeline_lr.named_steps['clr'].coef_[0]
````

````python
coeficientes_lr = pipeline_lr.named_steps['clr'].coef_[0]
````

````python
#Top 10 indicadores de NEGATIVIDAD
np.argsort(coeficientes_lr)[:10]
````

````python
#Top 10 indicadores de POSITIVIDAD
np.argsort(coeficientes_lr)[-10:]
````

````python
features_names_lr = pipeline_lr.named_steps['tfidf'].get_feature_names_out()
features_names_lr
````

````python
features_names_lr[np.argsort(coeficientes_lr)[:10]] #Palabras mas negativas
````

````python
features_names_lr[np.argsort(coeficientes_lr)[-10:]] #Palabras mas positivas
````

````python
print(classification_report(y_test, y_hat_test, target_names = ['Neg', 'Pos']))
````

### Support Vector Machine (SVM)

Las **SVM** encuentran el **hiperplano de máximo margen** que separa las clases.

Para texto se usa típicamente un **kernel lineal** con TF-IDF, lo que resulta en uno de los clasificadores más potentes para NLP clásico.

$$\text{Clasificar: } f(x) = \text{sign}(w^T x + b)$$

**Ventajas:**
- Muy efectivo en espacios de alta dimensión (como TF-IDF)
- Robusto al overfitting
- El parámetro `C` controla el trade-off margen/errores

````python
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
````

````python
#Pipeline
pipeline_svm = Pipeline([('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df = 1, max_features=5000, sublinear_tf = True)),
                        ('clsvm', CalibratedClassifierCV(LinearSVC(C = 1.0, max_iter = 2000)))
    
]) 
````

````python
#Entrenamiento
pipeline_svm.fit(X_train, y_train)
# SVM Lineal 
````

````python
y_pred_svm = pipeline_svm.predict(X_test)
````

````python
y_pred_svm
````

````python
print(classification_report(y_test, y_pred_svm, target_names = ['Neg', 'Pos']))
````

## Clasificación Multi-clase

Pasamos de binario (pos/neg) a **múltiples categorías**. Usaremos un dataset de noticias con 4 categorías.

````python
from sklearn.datasets import fetch_20newsgroups
from sklearn.preprocessing import LabelEncoder
from collections import Counter

# Subcategoría de 4 grupos temáticos
categorias = ['sci.space', 'rec.sport.baseball', 'talk.politics.guns', 'rec.autos']
news = fetch_20newsgroups(subset='all', categories=categorias,
                          remove=('headers', 'footers', 'quotes'))

print(f"Dataset: {len(news.data)} documentos, {len(news.target_names)} clases")
print(f"Clases: {news.target_names}")
````

````python
news.target_names
````

````python
# Distribución por clase
dist = Counter(news.target_names[t] for t in news.target)
print("\nDistribución:")
for clase, cnt in dist.most_common():
    print(f"  {clase}: {cnt}")
````

````python
# División train test split
X_tr, X_te, y_tr, y_te = train_test_split(
    news.data, news.target, test_size=0.2, random_state=42, stratify=news.target)
````

````python
# Comparar los 3 modelos en multi-clase
modelos = { 'Naive_Bayes': Pipeline([('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df = 1, max_features=5000, sublinear_tf = True)), ('clf', MultinomialNB(alpha=0.5))]),
            'Log_Reg': Pipeline([('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df = 1, max_features=5000, sublinear_tf = True)), ('clr', LogisticRegression(C=5, max_iter=1000))]),
            'svm': Pipeline([('tfidf', TfidfVectorizer(ngram_range=(1,2), min_df = 1, max_features=5000, sublinear_tf = True)),('clsvm', CalibratedClassifierCV(LinearSVC(C = 1.0, max_iter = 2000)))])
}
````

````python
resultados = {}
for mod, pipeline_mod in modelos.items():
    pipeline_mod.fit(X_tr, y_tr)
    y_p_test = pipeline_mod.predict(X_te)
    acc = accuracy_score(y_te, y_p_test)
    resultados[mod]  = acc
    print(f'\n ====== {mod}, accuracy: {acc} ==================')
    print(classification_report(y_te, y_p_test, target_names =news.target_names))
    
````

#### Métricas de Evaluación

| Métrica | Fórmula | Cuándo usar |
|---|---|---|
| **Accuracy** | $\frac{TP+TN}{TP+TN+FP+FN}$ | Clases balanceadas |
| **Precision** | $\frac{TP}{TP+FP}$ | Cuando los FP son costosos |
| **Recall** | $\frac{TP}{TP+FN}$ | Cuando los FN son costosos |
| **F1-Score** | $\frac{2 \cdot P \cdot R}{P+R}$ | Balance precision/recall |
| **AUC-ROC** | Área bajo curva ROC | Evaluación global del clasificador |

````python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')
from sklearn.metrics import ConfusionMatrixDisplay, roc_curve, auc
from sklearn.preprocessing import label_binarize

#  Comparación de modelos 
print("   Resumen Comparativo   ")
print(f"{'Modelo':<15} {'Accuracy':>10}")
print("-" * 28)
for nombre, acc in sorted(resultados.items(), key=lambda x: -x[1]):
    bar = '||' * int(acc * 30)
    print(f"{nombre:<15} {acc:>8.3f}  {bar}")
````
